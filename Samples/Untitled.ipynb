```python
p_list = np.array(torch.load('{}_p24.pth'.format(model_label)))
ph_list = np.array(torch.load('{}_ph24.pth'.format(model_label)))
```

```python
def amp_to_db(x):
#     print('amp', x[:,0])
    return 20*np.log10(x+1e-20)
output_l = []
for i in zip(p_list, ph_list):
    p = i[0]
    p_h = i[1]
    
    scale = np.mean(p)/np.mean(p_h)
    p_h *= scale
    print(p, p_h)
    output = abs(amp_to_db((p+1e-20)/(p_h+1e-20)))
    output_l.append(output)
    
#     if i == 3:
    break
```
